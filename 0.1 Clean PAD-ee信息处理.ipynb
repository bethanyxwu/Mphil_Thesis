1. 首先，处理一下ee_name及ee_information

```python
# 计算exec_dt列中缺失值的个数
missing_values_count = reassign_df['ee_name'].isna().sum()
missing_values_count
```

```python
# 对一个字段操作前的基本步骤：
# 指定的列
cols = ['ee_country', 'ee_state', 'ee_city', 'ee_postcode', 'ee_address_1', 'ee_address_2']

# 填充缺失值为空字符串，转换为字符串类型，并去除首尾空格
reassign_df[cols] = reassign_df[cols].fillna('').astype(str).apply(lambda x: x.str.strip())

reassign_df[cols].head()
```

```python
#1））补充ee_country；
import numpy as np
# 新建ee_country_full列，首先复制ee_country的值
reassign_df['ee_country_full'] = reassign_df['ee_country']

# 如果ee_country为空字符串且ee_state不为空字符串，填充"UNITED STATES"
reassign_df['ee_country_full'] = np.where(
    (reassign_df['ee_country_full'] == '') & (reassign_df['ee_state'] != ''),
    "UNITED STATES",
    reassign_df['ee_country_full']
)

# 替换"VIRGIN ISLANDS, U.S."为"UNITED STATES"
reassign_df['ee_country_full'] = reassign_df['ee_country_full'].replace("VIRGIN ISLANDS, U.S.", "UNITED STATES")
reassign_df[['ee_country_full','ee_country','ee_state','ee_city','ee_address_1','ee_address_2']].head(10)
```

```python
# 计算 ee_country_full 列中空字符串的数量
empty_string_count = (reassign_df['ee_country_full'] == '').sum()

# 打印结果
print(f"Number of empty strings in 'ee_country_full': {empty_string_count}")
```
