```python
import pandas as pd
documentid_path = '/content/drive/MyDrive/PAD_csv/documentid.csv'
documentid_df = pd.read_csv(documentid_path)
documentid_df
```

```python
# 删除rf_id列中的缺失值
documentid_df.dropna(subset=['rf_id'], inplace=True)
# 转换rf_id列为字符串，并去除可能的前后空格
documentid_df['rf_id'] = documentid_df['rf_id'].astype(str).str.strip()
# 删除空字符串的样本
documentid_df = documentid_df[documentid_df['rf_id'] != '']
```

```python
# 删除rf_id列中的缺失值
documentid_df.dropna(subset=['grant_doc_num'], inplace=True)
# 转换rf_id列为字符串，并去除可能的前后空格
documentid_df['grant_doc_num'] = documentid_df['grant_doc_num'].astype(str).str.strip()
# 删除空字符串的样本
documentid_df = documentid_df[documentid_df['grant_doc_num'] != '']
```

```python
import pandas as pd

# 使用 groupby 和 nunique 函数计算每个 rf_id 的不同 grant_doc_num 的数量
unique_grant_counts = documentid_df.groupby('rf_id')['grant_doc_num'].nunique().reset_index()

# 重命名列以更好地反映信息
unique_grant_counts.rename(columns={'grant_doc_num': 'unique_grant_count'}, inplace=True)

# 筛选 unique_grant_count 大于 25 的 rf_id
rf_ids_over_25 = unique_grant_counts[unique_grant_counts['unique_grant_count'] >= 25]['rf_id'].tolist()

print(len(rf_ids_over_25))
```

```python
import pandas as pd

# 定义文件路径
file_path = '/content/drive/MyDrive/reassignment/reassignment_eeFirm_invorScore_eeAssigneeScore_oreeScore_MA.csv'

# 使用read_csv加载指定的列
df = pd.read_csv(file_path)
print(len(df))
```

```python
# 在df中增加is_25MA列，根据rf_id是否在rf_ids_over_25列表中来赋值
df['is_25MA'] = df['rf_id'].isin(rf_ids_over_25).astype(int)
```

```python
# 计算is_25MA为1的样本数
count_is_25MA = df['is_25MA'].sum()

# 计算总样本数
total_samples = len(df)

# 计算is_25MA为1的比例
proportion_is_25MA = count_is_25MA / total_samples

# 打印结果
print(f"样本中 'is_25MA' 为 1 的数量: {count_is_25MA}")
print(f"总样本数: {total_samples}")
print(f"'is_25MA' 为 1 的比例: {proportion_is_25MA:.2f}")
```

这也说明之前的筛选已经很干净了，排除掉了基本可能的MA情况
