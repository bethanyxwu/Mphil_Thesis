```python
# 确保 df 中的 'fyear' 包含 2010 到 2023 年的所有样本
required_years = set(range(1997, 2023 + 1))

# 定义一个函数来检查每个 gvkey 是否包含所有必需的年份
def check_years(group):
    gvkey = group.name
    years = set(group['fyear'])
    missing_years = required_years - years
    return pd.Series({
        'gvkey': gvkey,
        'has_complete_years': len(missing_years) == 0,
        'missing_years': missing_years
    })

# 应用函数到每个 gvkey 分组
result = df.groupby('gvkey').apply(check_years).reset_index(drop=True)

# 打印不完整年份的 gvkey 以及缺失的年份
incomplete_gvkeys = result[result['has_complete_years'] == False]
print("不完整年份的 gvkey 以及缺失的年份:")
print(incomplete_gvkeys)
```

```python
import pandas as pd
from tqdm import tqdm

print("处理唯一值...")
# 添加进度条显示
tqdm.pandas()

# 处理唯一值的函数
def process_unique_values(row):
    if pd.isna(row['all_incumbent_CTO']):
        row['all_incumbent_CTO_unique'] = pd.NA
        row['YearStartRole_unique'] = pd.NA
        row['YearEndRole_unique'] = pd.NA
    else:
        director_ids = row['all_incumbent_CTO'].split('; ')
        start_years = row['YearStartRole'].split('; ')
        end_years = row['YearEndRole'].split('; ')

        unique_director_ids = []
        unique_start_years = []
        unique_end_years = []

        id_year_map = {}

        for director_id, start_year, end_year in zip(director_ids, start_years, end_years):
            if director_id not in id_year_map:
                id_year_map[director_id] = {'start_year': start_year, 'end_year': end_year}
            else:
                if start_year < id_year_map[director_id]['start_year']:
                    id_year_map[director_id]['start_year'] = start_year
                if end_year > id_year_map[director_id]['end_year']:
                    id_year_map[director_id]['end_year'] = end_year

        for director_id, years in id_year_map.items():
            unique_director_ids.append(director_id)
            unique_start_years.append(years['start_year'])
            unique_end_years.append(years['end_year'])

        row['all_incumbent_CTO_unique'] = '; '.join(unique_director_ids)
        row['YearStartRole_unique'] = '; '.join(unique_start_years)
        row['YearEndRole_unique'] = '; '.join(unique_end_years)

    return row

# 使用带进度条的apply函数
final_dataset = final_dataset.progress_apply(process_unique_values, axis=1)

print("处理完成！")

# 打印最终的数据集
final_dataset[['unique_CompanyID', 'fyear', 'all_incumbent_CTO_unique', 'YearStartRole_unique', 'YearEndRole_unique']]
```

```python
import pandas as pd
from tqdm import tqdm

# 创建从 DirectorID 到 AwardDate 的字典
phd_dates = dataset3.set_index('DirectorID')['AwardYear'].to_dict()

# 定义一个函数来检查每个 DirectorID 是否在 exec_dt 或之前获得了 PhD
def check_phd(incumbent_ids, exec_date):
    if incumbent_ids == '':  # 如果 incumbent_ids 是空字符串
        return ''  # 返回空字符串
    ids = incumbent_ids.split('; ')
    phd_status = []
    for id_ in ids:
        if id_ in phd_dates and phd_dates[id_] <= exec_date:
            phd_status.append('1')
        else:
            phd_status.append('0')
    return '; '.join(phd_status)

# 使用 tqdm 来显示进度条，应用函数到 df 的每一行
tqdm.pandas(desc="Checking PhD status")
df['is_CTO_phd'] = df.progress_apply(lambda x: check_phd(x['all_incumbent_CTO_unique'], x['fyear']), axis=1)

df[['all_incumbent_CTO_unique', 'fyear', 'is_CTO_phd']]
```

```python
import pandas as pd
from tqdm import tqdm

# 将 fyear 转换为整数格式
df['fyear'] = df['fyear'].astype(int)

# 对 df 按 'gvkey' 和 'fyear' 排序
df = df.sort_values(by=['gvkey', 'fyear'])

# 新增 'is_new_directorID' 和 'new_directorID' 列，初始值为 0 和 None
df['is_new_CTO'] = 0
df['new_CTOID'] = None

# 定义一个函数来找到新的 DirectorID 并去除重复值
def find_new_directors(curr_ids, prev_ids):
    curr_set = set(curr_ids.split('; ')) if pd.notna(curr_ids) and curr_ids.strip() != '' else set()
    prev_set = set(prev_ids.split('; ')) if pd.notna(prev_ids) and prev_ids.strip() != '' else set()
    new_directors = curr_set - prev_set
    return '; '.join(sorted(new_directors)) if new_directors else None

# 获取 gvkey 的数量用于进度条
total_gvkeys = df['gvkey'].nunique()

# 对每个 gvkey 分组进行操作，并显示进度条
for gvkey, group in tqdm(df.groupby('gvkey'), total=total_gvkeys, desc="Processing gvkeys"):
    # 按 fyear 排序
    group = group.sort_values(by='fyear')
    prev_ids = None
    prev_year = None

    for idx, row in group.iterrows():
        curr_year = row['fyear']
        curr_ids = row['all_incumbent_CTO_unique']

        # 只在年份连续的情况下进行比较
        if prev_year is not None and curr_year == prev_year + 1:
            if pd.notna(curr_ids) and curr_ids.strip() != '':
                if prev_ids is None or prev_ids.strip() == '':
                    # 上一年为空值，而这一年不为空值
                    df.at[idx, 'is_new_CTO'] = 1
                    df.at[idx, 'new_CTOID'] = curr_ids
                else:
                    # 检查是否有新的 DirectorID
                    new_directors = find_new_directors(curr_ids, prev_ids)
                    if new_directors:
                        df.at[idx, 'is_new_CTO'] = 1
                        df.at[idx, 'new_CTOID'] = new_directors

        prev_ids = curr_ids
        prev_year = curr_year

# 检查new_directorID列中的重复值并去除
df['new_CTOID'] = df['new_CTOID'].apply(lambda x: '; '.join(sorted(set(x.split('; ')))) if pd.notna(x) and x.strip() != '' else x)

df[['gvkey', 'fyear', 'all_incumbent_CTO_unique', 'is_new_CTO', 'new_CTOID']].head()
```

```python
import pandas as pd
from tqdm import tqdm

# 创建从 DirectorID 到 AwardDate 的字典
phd_dates = dataset3.set_index('DirectorID')['AwardYear'].to_dict()

# 定义一个函数来检查每个 DirectorID 是否在 exec_dt 或之前获得了 PhD
def check_phd(incumbent_ids, exec_date):
    if incumbent_ids == '':  # 如果 incumbent_ids 是空字符串
        return ''  # 返回空字符串
    ids = incumbent_ids.split('; ')
    phd_status = []
    for id_ in ids:
        if id_ in phd_dates and phd_dates[id_] <= exec_date:
            phd_status.append('1')
        else:
            phd_status.append('0')
    return '; '.join(phd_status)

# 使用 tqdm 来显示进度条，应用函数到 df 的每一行
tqdm.pandas(desc="Checking PhD status")
df['is_NewCTO_phd'] = df.progress_apply(lambda x: check_phd(x['new_CTOID'], x['fyear']), axis=1)

df[['gvkey', 'fyear', 'all_incumbent_CTO_unique', 'is_new_CTO', 'new_CTOID', 'is_NewCTO_phd']].head()
```
