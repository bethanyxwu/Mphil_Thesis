```python
import pandas as pd
from datetime import datetime

# 检查 exec_dt 列中是否存在无效日期格式
def is_valid_date(date_str):
    try:
        datetime.strptime(date_str, '%Y-%m-%d')
        return True
    except ValueError:
        return False

df['is_valid_exec_dt'] = df['exec_dt'].apply(is_valid_date)
invalid_dates = df[~df['is_valid_exec_dt']]

# 提取 exec_dt 的年份，存储在新列 exec_year 中
def extract_year(date_str):
    try:
        return datetime.strptime(date_str, '%Y-%m-%d').year
    except ValueError:
        return None

df['exec_year'] = df['exec_dt'].apply(extract_year)

print("无效日期样本如下:")
print(invalid_dates)
```

```python
df['gvkey'] = df['gvkey'].astype(str).str.strip()
df['gvkey']
```

```python
# 查找 gvkey, exec_year, grant_doc_num 都相同的样本
duplicates = df[df.duplicated(subset=['gvkey', 'exec_year', 'grant_doc_num'], keep=False)]

# 按顺序列出
duplicates_sorted = duplicates.sort_values(by=['gvkey', 'exec_year', 'grant_doc_num'])

print("重复样本如下:")
duplicates_sorted[['gvkey','exec_year','grant_doc_num','exec_dt','rf_id']]
```
