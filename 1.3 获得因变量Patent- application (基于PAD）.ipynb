```python
import pandas as pd

# 定义文件路径
file_path = '/Users/wuxinyu/Library/CloudStorage/OneDrive-HKUST(Guangzhou)/patent/addition_list/ApplicationInformation_initial.csv'

# 读取CSV文件
df = pd.read_csv(file_path)
df
```

```python
df[['gvkey','filing_year','grant_doc_num']]
```

```python
# 检查'gvkey', 'filing_year', 'patent_num'列是否存在缺失值
missing_values = df[['gvkey', 'filing_year', 'grant_doc_num']].isnull().sum()

print(missing_values)
```

```python
# 检查'gvkey', 'filing_year', 'patent_num'列是否存在重复样本
duplicates = df[df.duplicated(subset=['gvkey', 'filing_year', 'grant_doc_num'], keep=False)]

print(f"Number of duplicate rows: {len(duplicates)}")
print(duplicates)
```

```python
# 重命名'filing_year'为'year'
df = df.rename(columns={'filing_year': 'year'})
```

```python
df['year']
```

### 1. 首先统计Application的number - gvkey在year申请专利的数量

```python
#1. 首先统计Application的number - gvkey在year申请专利的数量

# 统计每个 gvkey, exec_year 组合出现的次数，并新建数据框
Apply_aggregated = df.groupby(['gvkey', 'year']).size().reset_index(name='number_Application')

Apply_aggregated
```

### 合并回已经获得了的自变量的Panel表

```python
import pandas as pd

# 读取Excel文件中的数据
panel_path = '/Users/wuxinyu/Library/CloudStorage/OneDrive-HKUST(Guangzhou)/毕业论文-Patent/data/used/benchmark_innovation_Updated.xlsx'
panel_df = pd.read_excel(panel_path)
panel_df
```

```python
panel_df['gvkey'] = panel_df['gvkey'].astype(str).str.strip()
panel_df['fyear'] = panel_df['fyear'].astype(int)
```

```python
panel_df[['gvkey','fyear']]
```

```python
Apply_aggregated['gvkey'] = Apply_aggregated['gvkey'].astype(str).str.strip()
Apply_aggregated['year'] = Apply_aggregated['year'].astype(int)
Apply_aggregated[['gvkey','year']]
```

```python
# 增加一列 is_MFTevent，全部赋值为 1
Apply_aggregated['is_Apply'] = 1
Apply_aggregated
```

```python
# 以dataset1为基准表，进行左合并
merged_df1 = pd.merge(panel_df, Apply_aggregated, how='left', left_on=['gvkey', 'fyear'], right_on=['gvkey', 'year'])

# 将 is_MFTevent 和 number_MFTevent 转为整数格式，忽略缺失值
merged_df1['is_Apply'] = merged_df1['is_Apply'].fillna(0).astype(int)
merged_df1['number_Application'] = merged_df1['number_Application'].fillna(0).astype(int)

# 删除 exec_year 列
merged_df1 = merged_df1.drop(columns=['year'])

merged_df1[['gvkey','fyear','is_Apply','number_Application']]
```

```python
# 计算 is_MFTevent 为 1 的个数
count_is_MFTevent_1 = merged_df1['is_Apply'].sum()

# 计算 number_MFTevent 列大于等于 1 的个数
count_number_MFTevent_ge_1 = (merged_df1['number_Application'] >= 1).sum()

print("is_Apply 为 1 的个数:", count_is_MFTevent_1)
print("number_Application 列大于等于 1 的个数:", count_number_MFTevent_ge_1)
```

```python
# 增加一列 is_MFTevent，全部赋值为 1
MFTevent_aggregated_utility['is_ApplyUtility'] = 1
MFTevent_aggregated_utility
```

```python
# 以dataset1为基准表，进行左合并
merged_df2 = pd.merge(merged_df1, MFTevent_aggregated_utility, how='left', left_on=['gvkey', 'fyear'], right_on=['gvkey', 'year'])

# 将 is_MFTevent 和 number_MFTevent 转为整数格式，忽略缺失值
merged_df2['is_ApplyUtility'] = merged_df2['is_ApplyUtility'].fillna(0).astype(int)
merged_df2['number_ApplyUtility'] = merged_df2['number_ApplyUtility'].fillna(0).astype(int)
merged_df2 = merged_df2.drop(columns=['year'])
merged_df2[['gvkey','fyear','is_Apply','number_Application','is_ApplyUtility','number_ApplyUtility']]
```

```python
# 计算 is_MFTevent 为 1 的个数
count_is_MFTevent_1 = merged_df2['is_ApplyUtility'].sum()

# 计算 number_MFTevent 列大于等于 1 的个数
count_number_MFTevent_ge_1 = (merged_df2['number_ApplyUtility'] >= 1).sum()

print("is_MFTevent 为 1 的个数:", count_is_MFTevent_1)
print("number_MFTevent 列大于等于 1 的个数:", count_number_MFTevent_ge_1)
```

```python
# 以dataset1为基准表，进行左合并
merged_df3 = pd.merge(merged_df2, ScienceUtility_aggregated, how='left', left_on=['gvkey', 'fyear'], right_on=['gvkey', 'year'])

# 将 is_MFTevent 和 number_MFTevent 转为整数格式，忽略缺失值
merged_df3['is_ScienceApply'] = merged_df3['is_ScienceApply'].fillna(0).astype(int)
merged_df3['number_ScienceApply'] = merged_df3['number_ScienceApply'].fillna(0).astype(int)
merged_df3 = merged_df3.drop(columns=['year'])
merged_df3[['gvkey','fyear','is_Apply','number_Application','is_ApplyUtility','number_ApplyUtility','is_ScienceApply','number_ScienceApply']]
```
