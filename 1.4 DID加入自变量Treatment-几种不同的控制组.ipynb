```python
import pandas as pd
from tqdm import tqdm

# 创建从 DirectorID 到 AwardDate 的字典
phd_dates = dataset3.set_index('DirectorID')['AwardYear'].to_dict()

# 定义一个函数来检查每个 DirectorID 是否在 exec_dt 或之前获得了 PhD
def check_phd(incumbent_ids, exec_date):
    if incumbent_ids == '':  # 如果 incumbent_ids 是空字符串
        return ''  # 返回空字符串
    ids = incumbent_ids.split('; ')
    phd_status = []
    for id_ in ids:
        if id_ in phd_dates and phd_dates[id_] <= exec_date:
            phd_status.append('1')
        else:
            phd_status.append('0')
    return '; '.join(phd_status)

# 使用 tqdm 来显示进度条，应用函数到 df 的每一行
tqdm.pandas(desc="Checking PhD status")
df['is_CTO_正常phd'] = df.progress_apply(lambda x: check_phd(x['all_incumbent_CTO_unique'], x['fyear']), axis=1)

df[['all_incumbent_CTO_unique', 'fyear', 'is_CTO_正常phd']]
```

```python
import pandas as pd
from tqdm import tqdm

# 将 fyear 转换为整数格式
df['fyear'] = df['fyear'].astype(int)

# 对 df 按 'gvkey' 和 'fyear' 排序
df = df.sort_values(by=['gvkey', 'fyear'])

# 新增 'is_new_directorID' 和 'new_directorID' 列，初始值为 0 和 None
df['is_new_CTO'] = 0
df['new_CTOID'] = None

# 定义一个函数来找到新的 DirectorID 并去除重复值
def find_new_directors(curr_ids, prev_ids):
    curr_set = set(curr_ids.split('; ')) if pd.notna(curr_ids) and curr_ids.strip() != '' else set()
    prev_set = set(prev_ids.split('; ')) if pd.notna(prev_ids) and prev_ids.strip() != '' else set()
    new_directors = curr_set - prev_set
    return '; '.join(sorted(new_directors)) if new_directors else None

# 获取 gvkey 的数量用于进度条
total_gvkeys = df['gvkey'].nunique()

# 对每个 gvkey 分组进行操作，并显示进度条
for gvkey, group in tqdm(df.groupby('gvkey'), total=total_gvkeys, desc="Processing gvkeys"):
    # 按 fyear 排序
    group = group.sort_values(by='fyear')
    prev_ids = None
    prev_year = None

    for idx, row in group.iterrows():
        curr_year = row['fyear']
        curr_ids = row['all_incumbent_CTO_unique']

        # 只在年份连续的情况下进行比较
        if prev_year is not None and curr_year == prev_year + 1:
            if pd.notna(curr_ids) and curr_ids.strip() != '':
                if prev_ids is None or prev_ids.strip() == '':
                    # 上一年为空值，而这一年不为空值
                    df.at[idx, 'is_new_CTO'] = 1
                    df.at[idx, 'new_CTOID'] = curr_ids
                else:
                    # 检查是否有新的 DirectorID
                    new_directors = find_new_directors(curr_ids, prev_ids)
                    if new_directors:
                        df.at[idx, 'is_new_CTO'] = 1
                        df.at[idx, 'new_CTOID'] = new_directors

        prev_ids = curr_ids
        prev_year = curr_year

# 检查new_directorID列中的重复值并去除
df['new_CTOID'] = df['new_CTOID'].apply(lambda x: '; '.join(sorted(set(x.split('; ')))) if pd.notna(x) and x.strip() != '' else x)

df[['gvkey', 'fyear', 'all_incumbent_CTO_unique', 'is_new_CTO', 'new_CTOID']].head()
```

```python
import pandas as pd
import numpy as np
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

# 假设 df 已经创建并包含必要的列

# 1. 删除'sic'列为缺失值的样本
df = df.dropna(subset=['sic'])

# 生成一列'missing_rd'，如果'xrd'列为缺失值则为1，否则为0
df['missing_rd'] = df['xrd'].isna().astype(int)

# 2.1 对于'xrd'列的缺失值用0填充
df['xrd_zero'] = df['xrd'].fillna(0)

# 2.2 对于'xrd'列的缺失值用行业 sic 均值填充，如果进一步存在缺失值用全样本均值填充
df['xrd_sic_ave'] = df.groupby('sic')['xrd'].transform(lambda x: x.fillna(x.mean()))
df['xrd_sic_ave'] = df['xrd_sic_ave'].fillna(df['xrd'].mean())

# 2.3 对于'xrd'列的缺失值用每年 fyear 的行业 sic 均值填充，如果进一步存在缺失值用行业 sic 均值填充，如果进一步存在缺失值用全样本均值填充
df['xrd_fyearsic_ave'] = df.groupby(['fyear', 'sic'])['xrd'].transform(lambda x: x.fillna(x.mean()))
df['xrd_fyearsic_ave'] = df.groupby('sic')['xrd_fyearsic_ave'].transform(lambda x: x.fillna(x.mean()))
df['xrd_fyearsic_ave'] = df['xrd_fyearsic_ave'].fillna(df['xrd'].mean())

# 3. 对于指定列的缺失值用多重插补
cols_to_impute = ['scf', 'act', 'at', 'capx', 'ch', 'che', 'cogs', 'emp', 'lct', 'lt', 'ni', 'revt', 'sale', 'teq']

# 选择需要插补的列
data_to_impute = df[cols_to_impute]

# 定义插补器
imputer = IterativeImputer(max_iter=10, random_state=0)

# 执行插补
imputed_data = imputer.fit_transform(data_to_impute)

# 将插补后的数据放回原DataFrame中
df[cols_to_impute] = imputed_data

# 显示结果
df[['sic', 'xrd', 'missing_rd', 'xrd_zero', 'xrd_sic_ave', 'xrd_fyearsic_ave'] + cols_to_impute]
```
